```python
import pandas as pd

# Membaca dua file CSV
res18 = pd.read_excel('res18_model_sampled.xlsx')
res34 = pd.read_excel('res34_model_sampled.xlsx')

# Menampilkan 5 baris pertama dari masing-masing file
print("===== RESNET-18 MODEL =====")
display(res18.head())

print("\n===== RESNET-34 MODEL =====")
display(res34.head())
```

```python
import numpy as np
import pandas as pd

# --- langsung pakai semua kolom dari res18 & res34 ---
A = res18.to_numpy(dtype=float)  # (n1 x p)
B = res34.to_numpy(dtype=float)  # (n2 x p)

# (opsional tapi disarankan: pastikan jumlah kolom sama)
assert A.shape[1] == B.shape[1], "Jumlah kolom res18 dan res34 harus sama."

# --- rata-rata per grup & grand mean ---
meanA = A.mean(axis=0)
meanB = B.mean(axis=0)
MEAN  = np.vstack([meanA, meanB])
DATA  = np.vstack([A, B])
GM    = DATA.mean(axis=0)

n1, p = A.shape
n2, p = B.shape
n = [n1, n2]
g = 2

# --- BETWEEN: B = Σ n_k (μ_k - μ)(μ_k - μ)' ---
Between = np.zeros((p, p))
for k in range(g):
    mu_k = MEAN[k, :]
    diff_col = (mu_k - GM).reshape(-1, 1)      # (p x 1)
    Between += n[k] * (diff_col @ diff_col.T)  # (p x p)

# --- WITHIN: W = Σ Σ (x_kj - μ_k)(x_kj - μ_k)' ---
Within = np.zeros((p, p))
for k in range(g):
    Xk  = A if k == 0 else B
    mu_k = MEAN[k, :]
    for j in range(Xk.shape[0]):
        diff_col = (Xk[j, :] - mu_k).reshape(-1, 1)
        Within  += diff_col @ diff_col.T

# --- TOTAL & WILKS' LAMBDA ---
Total  = Within + Between
Lambda = np.linalg.det(Within) / np.linalg.det(Total)

print("Mean A:", meanA)
print("Mean B:", meanB)
print("Grand Mean:", GM)
print("\nBetween:\n", Between)
print("\nWithin:\n", Within)
print("\nTotal:\n", Total)
print("\nWilks' Lambda:", Lambda)
```

```python
from scipy.stats import f

N_total = n1 + n2
df1 = p
df2 = N_total - p - 1

F_value = ((1 - Lambda) / Lambda) * ((N_total - p - 1) / p)


alpha = 0.05
F_crit = f.isf(alpha, df1, df2)

print(f"df1 = {df1}, df2 = {df2}")
print(f"F hitung = {F_value:.6f}")
print(f"F kritis (alpha={alpha}) = {F_crit:.6f}")


if F_value > F_crit:
    print("Keputusan: Tolak H0 (ada perbedaan signifikan antar 2 grup).")
else:
    print("Keputusan: Gagal menolak H0 (tidak ada perbedaan signifikan).")
```
